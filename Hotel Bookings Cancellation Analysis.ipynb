```python
# Number of cancelled and non-cancelled bookings
cancelled = df['is_canceled'].value_counts()
cancelled
```

```python
# Creating a bar chart of cancelled and non-cancalled bookings
plt.figure(figsize = (6,4))
plt.title('Reservation Status Count', size = 15)
plt.bar(['Not_Cancelled', 'Cancelled'], df['is_canceled'].value_counts())
plt.ylabel('Number of Reservations')

# Display the chart
plt.show()
```

```python
# Percentage of cancelled and non-cancelled bookings
perc_cancelled = df['is_canceled'].value_counts(normalize = True)
perc_cancelled
```

```python
# Creating a plot of cancelled and non-cancelled bookings in different hotels
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df)
legend_labels,_= ax1.get_legend_handles_labels()
plt.title('Reservation Status in Different Hotels', size = 15)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')

# Display the chart
plt.show()
```

```python
# Percentage of cancelled and non-cancelled bookings in Resort Hotels
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Percentage of cancelled and non-cancelled bookings in City Hotels
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Checking the Average Daily Rates in City and Resort Hotels
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Creating a plot of Average Daily Rates on City and Resort Hotels
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 15)

# Display the chart
plt.show()
```
